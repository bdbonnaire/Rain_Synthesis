```python
from IPython.display import Audio, display
import numpy as np
from Sound import Sound
```

```python
base_rain = Sound("./Sounds/Rain/heavy_rain_00_mono_short.wav")
base_rain.play()
```

```python
import pywt
```

```python
# These options follow the paper.
wavelet = 'db4'
level = 5

dwt_base_sound = pywt.wavedec(base_rain._array, wavelet, level=level)

dwt_base_sound[0] *= 2
out_array = pywt.waverec(dwt_base_sound, wavelet=wavelet)
modified_rain = Sound.from_array(out_array)
modified_rain.play()
```

It does seem to work well to give a sense of harder rain. Would need some kind of threshold to avoid the artifacts. Maybe some smoothing ?
